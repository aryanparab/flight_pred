```python
def preprocessing(df_all):
    df_all['Date'] = df_all['Date_of_Journey'].str.split('/').str[0]
    df_all['Month'] = df_all['Date_of_Journey'].str.split('/').str[1]
    df_all['Year'] = df_all['Date_of_Journey'].str.split('/').str[2]

    df_all['Date'] = df_all['Date'].astype(int)
    df_all['Month'] = df_all['Month'].astype(int)
    df_all['Year'] = df_all['Year'].astype(int)

    df_all.drop(['Additional_Info','Route'],axis=1,inplace=True)

    df_all['Arrival_Time'] = df_all['Arrival_Time'].str.split().str[0]
    df_all['Arrival_hour'] = df_all['Arrival_Time'].str.split(':').str[0]
    df_all['Arrival_min'] = df_all['Arrival_Time'].str.split(':').str[1]
    df_all['dep_hour'] = df_all['Dep_Time'].str.split(':').str[0]
    df_all['dep_min'] = df_all['Dep_Time'].str.split(':').str[1]

    df_all.drop(['Arrival_Time','Dep_Time'],axis=1,inplace = True)

    duration = list(df_all["Duration"])
    for i in range(len(duration)):
        if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
            if "h" in duration[i]:
                duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
            else:
                duration[i] = "0h " + duration[i]           # Adds 0 hour
    duration_hours = []
    duration_mins = []
    for i in range(len(duration)):
        duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
        duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))  

    df_all["Duration_hours"] = duration_hours
    df_all["Duration_mins"] = duration_mins

    df_all['Total_Stops'].replace({'non-stop':'0 stops'},inplace = True)
    df_all['Total_Stops']= df_all['Total_Stops'].fillna('1 stop')
    df_all['Stops'] =df_all['Total_Stops'].str.split(' ').str[0]
    df_all['Stops'] = df_all['Stops'].astype(int)

    df_all.drop('Total_Stops',axis=1,inplace = True)
    df_all.drop('Date_of_Journey',axis=1,inplace = True)

    df_airline = pd.get_dummies(df_all['Airline'])
    df_source = pd.get_dummies(df_all['Source'])
    df_dest = pd.get_dummies(df_all['Destination'])

    df_train = pd.concat((df_all,df_airline,df_source,df_dest),axis=1)
    df_train.drop(['Airline','Source','Destination','Duration'],axis=1,inplace = True)
    
    return df_train
```
